```python
from sklearn.linear_model import LinearRegression
?LinearRegression#gives detailed description and attributes of the linear regression class along with an example
```

```python
boston_df=DataFrame(boston.data)#boston.data me sara data aa jyega 
boston_df#column names are not there 
```

```python
boston_df.columns=boston.feature_names#it will set the column names 
boston_df
```

```python
boston_df['Price']=boston.target#it will store all the target data
boston_df.head(10)
```

```python
x=boston_df.RM
x.head()
```

```python

x=np.array([[val,1] for val in x])#in the RM column we iterate for each value #cant reshape as 506 are there
x=np.vstack(x)#converts it into 1 d array
print(x)
y=boston_df.Price
y=np.ravel(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
#x_train.shape
#y_train.shape
#x_test.shape
#y_test.shape
lreg = LinearRegression()#object of the linearregression class
lreg.fit(x_train,y_train)
#x_train.shape
#y_train.shape
```

```python
ypred=lreg.predict(x_test)
res=pd.concat([DataFrame(y_test,columns=['y_test']),DataFrame(ypred,columns=['ypred'])],axis=1)
lreg.score(x_test,y_test)
```

```python
res#comparing - actual price :ytest and predicted price:ypred
```

```python
#x=boston_df.CRIM+boston_df.AGE+boston_df.RAD+boston_df.RM+boston_df.INDUS
x=boston_df.drop(['Price'],axis=1)#we take all the features 
y=boston_df.Price
y=np.ravel(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

```python

from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=300,random_state = 0)#here as we increase the no. of estimators we get a better result and we get closer to the exact answer
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
res=pd.concat([DataFrame(y_test,columns=['y_test']),DataFrame(y_pred,columns=['y_pred'])],axis=1)
regressor.score(x_test,y_test)
```
